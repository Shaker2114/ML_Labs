**Считывание датасета**

```python
import pandas as pd
import matplotlib.pyplot as plt

dataframe = pd.read_csv('Students Performance.csv')

print(dataframe.head())
```

**3) Среднее значение, ско**

```python
mean_values = dataframe.mean() #среднее значение каждого атрибута
std_values = dataframe.std() #стандартное отклонение каждого атрибута
print(pd.concat([mean_values, std_values], axis=1, keys=['mean', 'std'])) #конкатенируем полученные фреймы по столбцам
```

**Гистограммы распределения значений, есть ли выбросы**

```python
# создание гистограммы
dataframe.hist(figsize=(11,11), edgecolor = 'black')
plt.show()
```

По полученным данным можно сказать, что выбросы не наблюдаются ни в одной из выше представленных гистограмм для каждого атрибута

**Пропущенные значения**

```python
# Пропущенные значения
missing_values = dataframe.isnull().sum()
print("Пропущенные значения:\n", missing_values)
```

Видим, что пропущенных значений в датасете нет

**4) Корреляция между параметрами**

```python
import seaborn as sns
corr_mat = dataframe.corr()
sns.heatmap(data = corr_mat, annot = True, cbar_kws={'label': 'Коэффицент корреляции'})
plt.show()
#сопоставление модулей коэффицентов корреляции с характеристикой корреляционной связи
#0.0-0.2 : отсутствует или практически отсутствует
#0.2-0.4 : слабая
#0.4-0.6 : средняя
#0.6-0.8 : высокая
#0.8-1.0 : очень высокая
```

**Матрица графиков рассеивания**

```python
sns.pairplot(dataframe)
plt.show()
```

**Вывод**

Мы видим, что корреляция между каждыми двумя атрибутами очень слабая, поэтому отсюда можно сделать вывод, что атрибуты практически не влияют друг на друга.
